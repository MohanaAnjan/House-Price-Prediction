**Hypothesis**

```python
#spliting category and continous
catagory=[]
continous=[]
threshold=15
for i in df.columns:
    if df[i].nunique()<threshold:
        catagory.append(i)
    else:
        continous.append(i)
```

```python
catagory,continous
```

```python
def two_sample(d1,d2): #continuous vs continuous
  m=[0,0]
  for i in range(31):
      sample1 = d1.sample(frac=0.03)
      sample2 = d2.sample(frac=0.04)
      t_test , p_value = stats.ttest_ind(sample1,sample2) #two_sample t-test
      if p_value < 0.03: #setting critical point , level of significance = 0.03
          m[1]+=1
      else:
          m[0]+=1

  if m[0]>m[1]:
    #print("H0 accept null hypothesis, Ha rejects alternate hypothesis, dependent samples , data is normally distributed")
    return True
  elif m[0]<m[1]:
    #print("H0 Reject null hypothesis, Ha accepts alternate hypothesis , independent samples , data is not normally distributed")
    return False
```

```python
def Chi_square(d1,d2): #category vs category
  value = stats.chi2_contingency(pd.crosstab(d1,d2).values)
  if value[1] < 0.03:
      #print("reject null hypothesis , accept alternative hypothesis , dependent")
      return True
  else:
      #print("accept null hypothesis , reject alternative hypothesis , independent")
      return False
```

```python
def Chi_square1(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2).values)[1] < 0.03 else False
```

```python
def annova(c1,c2): #category vs continuous
  group=df[c1].unique()
  data={}
  for i in group:
    data[i]=df[c2][df[c1]==i]
  f_value , p_value = stats.f_oneway(*[i for i in data.values()])
  if p_value < 0.05:
    #print("reject null hypothesis , accept alternative hypothesis , independent")
    return False
  else:
    # print("accept null hypothesis , reject alternative hypothesis , dependent")
    return True
```

```python
def annova1(c1,c2): #category vs continuous
  group=df[c1].unique()
  data={}
  for i in group:
    data[i]=df[c2][df[c1]==i]
  return False if stats.f_oneway(*[i for i in data.values()])[1] < 0.05 else True 
```

```python
house={}
for i in df.columns:
  house[i]={}
  for j in df.columns:
    if i=='SALES_PRICE' or j=='SALES_PRICE':
      continue
    elif i in continous and j in continous:
      result=two_sample(df[i],df[j])
    elif i in catagory and j in catagory:
      result=Chi_square(df[i],df[j])
    elif i in catagory and j in continous:
      result=annova(i,j)
    elif i in continous and j in catagory:
      result=annova(j,i)
    if result:
      house[i][j]=1
    else:
      house[i][j]=0
```

```python
house_df=pd.DataFrame(house)
house_df.drop("SALES_PRICE",axis=1,inplace=True)
house_df
```
